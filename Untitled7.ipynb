Crossings (likely pedestrian crossings) are much more frequently associated with a traffic signal when accidents occur â€” about 62% of crossing accidents happen where thereâ€™s a traffic signal.
This makes sense: crossings in high-traffic urban areas (where signals are common) see more pedestrian activity and possibly more incidents.
It could also reflect that signals donâ€™t entirely prevent accidents at crossings, or that crossings are riskier in dense areas.
Junctions see far fewer accidents involving a traffic signal â€” only 1.63% of junction accidents happen where thereâ€™s a traffic signal.
This might seem counterintuitive, but it likely reflects that:
Many junctions (especially rural or minor roads) donâ€™t have traffic signals.
Or that accidents at junctions with signals are relatively rare compared to those without signals, perhaps because signals are more effective at preventing junction collisions than crossing collisions.

# Linear regrssion of log(accident count) vs log(population count)


```python
X=np.log(Acc_and_pop_counts['Population count'])
y=np.log(Acc_and_pop_counts['Acc tot'])
X_const=sm.add_constant(X)
PopVsAcc=sm.OLS(y,X_const).fit()
y_pred=PopVsAcc.predict(X_const)
```

```python

plt.figure(figsize=(8,6))
plt.scatter(X,y,label='Observed data')
plt.plot(X,y_pred,color='red',label='Fitted line')
plt.title('log(Accidents) vs log(Population) across states OLS regression')
plt.legend()
plt.xlabel('log(Population)')
plt.ylabel('log(Accidents)')
```

## One outlier identified. 
The following describes regression stats with the outlier, and below that I will single out manually the outlier and show that regression is significantly improved.

```python
residuals = y_pred - y
plt.figure(figsize=(8,6))
plt.figure(figsize=(8,6))

# Histogram with KDE
sns.histplot(residuals, bins=6, color='skyblue')

plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Distribution of Residuals')
plt.axvline(x=0, color='black', linestyle='--')
plt.show()
```

```python
stat, p = shapiro(residuals)
if p > 0.05:
    print("===============================\n !!Probably normal according to Shapiro-Wilk test!!\n===============================\n")
else:
    print('===============================\n !!Probably NOT normal according to Shapiro-Wilk test!!\n===============================\n')
```

```python
print(PopVsAcc.summary())
```

# Removing the outlier
We can see in the scatter plot that the y-value of the outlier is $\leq$ 6. 

```python
outlier_states = Acc_and_pop_counts.loc[y[y <= 6].index, 'State']
print(f"The state associated with the outlier is {outlier_states.iloc[0]}")
```

```python

#Removing outlier

indx=y[y>6].index
X_rem=X[indx]
y_rem=y[indx]
X_rem_const=sm.add_constant(X_rem)
PopVsAcc_rem=sm.OLS(y_rem,X_rem_const).fit()
y_rem_pred=PopVsAcc.predict(X_rem_const)


plt.figure(figsize=(8,6))
plt.scatter(X_rem,y_rem,label='Observed data')
plt.plot(X_rem,y_rem_pred,color='red',label='Fitted line')
plt.title('log(Accidents) vs log(Population) across states OLS regression')
plt.legend()
plt.xlabel('log(Population)')
plt.ylabel('log(Accidents)')


residuals_rem = y_rem_pred - y_rem
plt.figure(figsize=(8,6))
plt.figure(figsize=(8,6))

# Histogram with KDE
sns.histplot(residuals_rem, bins=6, color='skyblue')

plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Distribution of Residuals')
plt.axvline(x=0, color='black', linestyle='--')
plt.show()

stat, p = shapiro(residuals_rem)
if p > 0.05:
    print("===============================\n !!Probably normal according to Shapiro-Wilk test!!\n===============================\n")
else:
    print('===============================\n !!Probably NOT normal according to Shapiro-Wilk test!!\n===============================\n!!')

print(PopVsAcc_rem.summary())
```

```python
mean_residual = np.mean(residuals_rem)
std_residual = np.std(residuals_rem, ddof=1)  # ddof=1 for sample standard deviation
print(f"Residual's mean is {mean_residual} and the std is {std_residual}")
```

## Summary for log(accidents) vs log(population) relationship

The above analysis establishes that $ \text{accidents}\approx e^{-10.87}\cdot \text{population}^{1.41}\cdot e^\epsilon$
where $\epsilon\sim\text{N}(-0.063,0.689)$



```python
plt.figure(figsize=(15,10))
coeff_array = PopVsAcc_rem.params
a=coeff_array['Population count']
b=coeff_array['const']
x_fit=Acc_and_pop_counts['Population count'].copy().sort_values()
y_fit=(np.exp(b)*x_fit**a)
plt.plot(x_fit,y_fit,color='red')
plt.scatter(Acc_and_pop_counts['Population count'],Acc_and_pop_counts['Acc tot'],color='blue',alpha=0.7,s=6)
threshold = Acc_and_pop_counts['Acc tot'].quantile(0.75)
for i, row in Acc_and_pop_counts.iterrows():
    if row['Acc tot']>threshold:
        plt.text(row['Population count'], row['Acc tot'], row['State'], fontsize=11, alpha=1,
             ha='right', va='bottom')
plt.title('Total Number of Accidents vs. State Population',size=15)
plt.xlabel('Population count')
plt.ylabel('Accidents count')
```
